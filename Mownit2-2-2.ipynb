```python
dates = pd.to_datetime(data["utc_timestamp"])
```

```python
dates
```

```python
ndata = pd.DataFrame({'date' : dates, 'solar' : data["DE_solar_generation"]})
```

```python
ndata
```

```python
plt.plot(ndata["date"], ndata["solar"])
plt.show()
```

```python
gdata = ndata.groupby(pd.Grouper(key = 'date', freq = '1D')).mean()
```

```python
gdata
```

```python
plt.plot(gdata)
plt.show()
```

```python
gdata.isnull().sum()
```

```python
original_data = gdata.dropna()
```

```python
original_data
```

```python
plt.plot(original_data)
```

```python
clear_data = pd.DataFrame(original_data).copy()
```

```python
plt.plot(clear_data)
```

```python
for index, row in clear_data.iterrows():
    if pd.Timestamp.to_pydatetime(index) < datetime.datetime.strptime("2015-01-01", "%Y-%m-%d") and pd.Timestamp.to_pydatetime(index) > datetime.datetime.strptime("2014-01-01", "%Y-%m-%d"):
        row["solar"] = np.nan
```

```python
plt.plot(clear_data)
```

```python
clear_data.isnull().sum()
```

# Deletion

```python
deleted_data = clear_data.dropna()
```

```python
plt.plot(original_data)
plt.plot(deleted_data)
```

# Replace with mean

```python
mean_replace_data = clear_data.replace(np.nan, clear_data.mean())
```

```python
clear_data.mean()
```

```python
plt.plot(original_data)
plt.plot(mean_replace_data)
```

## Replace with median

```python
median_replace_data = clear_data.replace(np.nan, clear_data.median())
```

```python
clear_data.median()
```

```python
plt.plot(original_data)
plt.plot(median_replace_data)
```

## Replace with mode

```python
mode_replace_data = clear_data.replace(np.nan, clear_data["solar"].mode()[1])
```

```python
plt.plot(original_data)
plt.plot(mode_replace_data)
```
